# RSS情報からの「見出し」取得用スクレイピングコード

```python
import numpy as np
import pandas as pd
import schedule
import time
import feedparser
import urllib.request, urllib.error
import os, datetime
from bs4 import BeautifulSoup

pd.set_option("display.max_rows",70)
pd.set_option('display.max_columns', 70)
```

```python
# RSS_list1はヤフーニュース。RSS_list2はまとめサイトのRSS。

RSS_list1 =["https://news.yahoo.co.jp/pickup/entertainment/rss.xml",
            "https://news.yahoo.co.jp/pickup/rss.xml",
            "https://news.yahoo.co.jp/pickup/world/rss.xml",
            "https://news.yahoo.co.jp/pickup/local/rss.xml",
            "https://news.yahoo.co.jp/pickup/domestic/rss.xml",
            "https://news.yahoo.co.jp/pickup/economy/rss.xml",
            "https://news.yahoo.co.jp/pickup/sports/rss.xml",
            "https://news.yahoo.co.jp/pickup/science/rss.xml",
            "https://news.yahoo.co.jp/pickup/entertainment/rss.xml",
            "https://news.yahoo.co.jp/pickup/computer/rss.xml"]

RSS_list2 =["http://blog.livedoor.jp/nwknews/index.rdf",
             "http://blog.livedoor.jp/nanjstu/index.rdf",
            "http://blog.livedoor.jp/dqnplus/index.rdf",
            "http://military38.com/index.rdf"]
```

```python
def RSS_GET(list,DF):
    for rss in list:
        feed = feedparser.parse(rss)
        Date = feed.headers["Date"]
        for entry in feed.entries:
            temp= pd.Series([entry.title,entry.link, Date],index=DF.columns)
            if temp.url not in DF.url:
                DF = DF.append(temp,ignore_index=True)        
    return DF
```

```python
X1 = pd.DataFrame(columns=["title","url","date"])
X1= RSS_GET(RSS_list1,X1)
X2 = pd.DataFrame(columns=["title","url","date"])
X2 = RSS_GET(RSS_list2,X2)

now = datetime.datetime.now()
filename1 = 'yahoo_rss_{0:%Y%m%d}.csv'.format(now)
filename2 = 'matome_rss_{0:%Y%m%d}.csv'.format(now)

X1.to_csv(filename1,encoding='utf_8_sig',mode='w')
X2.to_csv(filename2,encoding='utf_8_sig',mode='w')
```
